# 📊 Seaborn Tutorial
Dieses Notebook bietet eine Einführung in die Datenvisualisierung mit [Seaborn](https://seaborn.pydata.org/), einer auf Matplotlib basierenden Bibliothek für statistische Grafiken.

## 🔧 Vorbereitung
Wir starten mit dem Import der benötigten Bibliotheken und laden einen Beispiel-Datensatz.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Stil setzen
sns.set_theme(style="whitegrid")

# Beispiel-Datensatz laden
df = sns.load_dataset("tips")
df.head()
```

## 📈 1. Scatterplot
Ein Streudiagramm zeigt die Beziehung zwischen zwei numerischen Variablen.

```python
# Beziehung zwischen Gesamtbetrag und Trinkgeld
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", style="smoker")
```

## 📊 2. Histogramm
Ein Histogramm zeigt die Verteilung einer einzelnen numerischen Variable.

```python
# Verteilung der Rechnungsbeträge
sns.histplot(data=df, x="total_bill", bins=20, kde=True)
```

## 📦 3. Boxplot
Ein Boxplot visualisiert Median, Quartile und Ausreißer.

```python
# Vergleich der Trinkgeldhöhe nach Geschlecht
sns.boxplot(data=df, x="sex", y="tip")
```

## 📐 4. Violinplot
Eine Kombination aus Boxplot und Dichteplot.

```python
# Violinplot für das Trinkgeld nach Tag
sns.violinplot(data=df, x="day", y="tip", hue="sex", split=True)
```

## 🔢 5. Heatmap
Zeigt Korrelationen oder Kreuztabellen als Farbraster.

```python
# Korrelation zwischen numerischen Variablen
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
```
