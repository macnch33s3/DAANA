## ðŸ”§ Vorbereitung
Wir starten mit dem Import der benÃ¶tigten Bibliotheken und laden einen Beispiel-Datensatz.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Stil setzen
sns.set_theme(style="whitegrid")

# Beispiel-Datensatz laden
df = sns.load_dataset("tips")
df.head()
```

## ðŸ“ˆ 1. Scatterplot
Ein Streudiagramm zeigt die Beziehung zwischen zwei numerischen Variablen.

```python
# Beziehung zwischen Gesamtbetrag und Trinkgeld
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", style="smoker")
```

## ðŸ“Š 2. Histogramm
Ein Histogramm zeigt die Verteilung einer einzelnen numerischen Variable.

```python
# Verteilung der RechnungsbetrÃ¤ge
sns.histplot(data=df, x="total_bill", bins=20, kde=True)
```

## ðŸ“¦ 3. Boxplot
Ein Boxplot visualisiert Median, Quartile und AusreiÃŸer.

```python
# Vergleich der TrinkgeldhÃ¶he nach Geschlecht
sns.boxplot(data=df, x="sex", y="tip")
```

## ðŸ“ 4. Violinplot
Eine Kombination aus Boxplot und Dichteplot.

```python
# Violinplot fÃ¼r das Trinkgeld nach Tag
sns.violinplot(data=df, x="day", y="tip", hue="sex", split=True)
```

## ðŸ”¢ 5. Heatmap
Zeigt Korrelationen oder Kreuztabellen als Farbraster.

```python
# Korrelation zwischen numerischen Variablen
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
```
